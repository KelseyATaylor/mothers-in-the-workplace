# Women in the Workplace Overtime Capstone

## Imports & Data Inspection / Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
%matplotlib inline
```

```python
df = pd.read_csv('Women_in_Workplace_Master.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df = df.drop([51,47]).reset_index(drop=True)
```

## EDA

```python
df.describe()
```

```python
earners = df[['year','husband_only', 'wife_only', 'husband_wife']].copy()
```

```python
earners.dropna(inplace=True)
```

```python
fig = sns.boxplot(data=earners)
plt.ylabel('Number of Earners')
```

```python
sns.lineplot(data=earners, x='year', y='husband_only', label='husband only')
sns.lineplot(data=earners, x='year', y='wife_only', label='wife only')
sns.lineplot(data=earners, x='year', y='husband_wife', label='Husband and Wife')

plt.ylabel("Number of Earners")
plt.show()
```

```python
cpi = df[['year','cpi_average']]
sns.lineplot(data=cpi, x='year', y='cpi_average')
```

```python
df.info()
```

```python
age = df[['year','emp_15_24', 'emp_25_34', 'emp_35_44', 'emp_45_54']]
```

```python
age.describe()
```

```python
plt.figure(figsize=(12,8))

fig = sns.lineplot(data=age, x='year', y='emp_15_24', lw=3, label='15 to 24')
fig = sns.lineplot(data=age, x='year', y='emp_25_34', lw=3, label='25 to 34')
fig = sns.lineplot(data=age, x='year', y='emp_35_44', lw=3, label='35 to 44')
fig = sns.lineplot(data=age, x='year', y='emp_45_54', lw=3, label='45 to 54')

plt.xlabel('Year')
plt.ylabel("Number of Employed Women")

plt.title('Age of women Employed')
```

```python
age_pc = df[['year', 'emp_per_capita_15_24', 'emp_per_capita_25_34', 'emp_per_capita_35_44', 'emp_per_capita_45_54']]
```

```python
plt.figure(figsize=(16,8))

fig = sns.lineplot(data=age_pc, x='year', y='emp_per_capita_15_24', lw=3, label='15 to 24')
fig = sns.lineplot(data=age_pc, x='year', y='emp_per_capita_25_34', lw=3, label='25 to 34')
fig = sns.lineplot(data=age_pc, x='year', y='emp_per_capita_35_44', lw=3, label='35 to 44')
fig = sns.lineplot(data=age_pc, x='year', y='emp_per_capita_45_54', lw=3, label='45 to 54')

plt.xlabel('Year')
plt.ylabel("Number of Employed Women")

plt.title('Age of women Employed')
```
